```python
import random
print(random.random() )
```

```python
from sklearn import datasets
iris=datasets.load_iris()
d=iris.data
d.shape    
```

```python
#sample 30 points from 150
n=150
m=30
p=m/n
sampled_data=[]
for i in range(0,n):
    if random.random()<=p:
        sampled_data.append(d[i,:])
len(sampled_data)
```
